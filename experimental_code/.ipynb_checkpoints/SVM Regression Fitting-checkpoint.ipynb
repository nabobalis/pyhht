Recently, I and a friend have been attempting to magic up an empirical mode decomposition algorithm. The edge effects from the spline fitting are one of the major problems of the algorithm. We attempted to do a mirror expansion of the signal but that idea only works if it is possible to mirror the start and end of the signal but that is not always the case.

I came across a paper titled, "Elimination of end effects in empirical mode decomposition by mirror image coupled with support vector regression" located [here](http://adsabs.harvard.edu/abs/2012MSSP...31...13L) or [here.](http://www.sciencedirect.com/science/article/pii/S0888327012000726) I can provide the PDF or any of the images if people can not access it directly. 

They use the following method (as they outlined in Fig.8):

1) They use the first 4 local extrema points closest to the end points as the learning data. They mention that 4-6 is a good number of extrema points to use.

2) Then they test the model by extrapolating 3 more local extrema points and compare it to the signal. 

3) Finally, once verified, they will extrapolate two exterma points beyond the endpoints.

4) Then you mirror this extrapolation so you can improve the edge effects when you spline fit.

They state that they use the radial kernel: K(t,t_i) = exp(0.5(t-t_i)^2 / sigma^2) where sigma > 0 

I have attempted to replicate the results using SVR within scikit-learn but I can't reproduce their results. I suspect its due to my ignorance on using scikit-learn.

```python
# Lovely Imports
%matplotlib inline
import numpy as np
from sklearn.svm import SVR
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
# One of the most convoluted ways to generate a signal.
time = np.linspace(0,500,500)
data = 50. * np.sin(2*np.pi*time/10.) + 100. * np.sin(2*np.pi*time/20.) + 10 * np.random.rand(time.shape[0])
```

```python
# Initialise the SVR with kernal radial function.
svr_rbf = SVR(kernel='rbf', C=1e5, gamma=0.1, cache_size=2500,verbose=True)
# To make sure the x axis have the correct shape.
X = time[:,None]
# Predict the signal forwards for comparision.
pos_pred = svr_rbf.fit(X[0:40], data[0:40]).predict(X[40:60])
# Predict the signal backwards.
neg_pred = svr_rbf.fit(X[0:40], data[0:40]).predict(-X[0:40])
```

```python
# Plot this!
plt.plot(-time[0:40],neg_pred, 'b-_')
plt.plot(time[40:60],pos_pred, 'g-_')
plt.plot(time,data, 'r-.')
plt.show()
```
