```python
from __future__ import division
import numpy as np
import GPy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
basis = np.linspace(0,100,100)
data = 2 * np.sin(2*np.pi*basis/10) + 3 * np.sin(2*np.pi*basis/20)
```

```python
kernel = GPy.kern.rbf(input_dim=1, variance=1)
m = GPy.models.GPRegression(basis[:,None],data[:,None],kernel)
```

```python
plt.plot(basis,data, 'r-.')
m.plot()
```
