```python
Recently, I and a friend have been attempting to magic up an Empirical Mode Decomposition (EMD) algorithm. The edge effects from the spline fitting are one of the major problems of the algorithm. We attempted to do a mirror expansion of the signal but that idea only works if it is possible to mirror the start and end of the signal but that is not always the case.

I came across a paper titled, "Elimination of end effects in empirical mode decomposition by mirror image coupled with support vector regression" located [here](http://adsabs.harvard.edu/abs/2012MSSP...31...13L) or [here.](http://www.sciencedirect.com/science/article/pii/S0888327012000726) I can provide the PDF or any of the images if people can not access it directly. 

They use the following method (as they outline in Fig.8):

1) They use the first 4 local extrema points closest to the end points as the learning data. They mention that 4-6 is a good number of extrema points to use.

2) Then they test the model by extrapolating beyond 3 more local extrema points and compare it to the signal. 

3) Finally, once verified, they will extrapolate two exterma points beyond the endpoints.

They use the radial kernel: $K(t,t_i) = \exp(0.5(t-t_i)^2/\sigma^2)$ where $\sigma > 0$.
"Where $\sigma$ is the kernel width parameter related to the confidence in the data and the non-linear nature of the regression. As
 $\sigma$ increases, the kernel becomes wider, forcing the model toward a less complex (more linear) solution." 

To my eyes their radial kernel function is different to the one you use ($\exp(-\gamma |x-x'|^2)$. $\gamma$ is specified by keyword gamma, must be greater than 0.). However, I am uncertain.

I have attempted to replicate the results using SVR within scikit-learn but I can't reproduce their results. I suspect its due to my ignorance on using scikit-learn.
```

```python
from __future__ import division
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import mpld3
%matplotlib inline
mpld3.enable_notebook()
```

```python
basis = np.linspace(0,100,100)
data = 2 * np.sin(2*np.pi*basis/10) + 3 * np.sin(2*np.pi*basis/20) + 10.*np.random.random(len(basis))
```

```python
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1)
X = np.linspace(0,data.size,data.size)[:,None]
neg_pred = svr_rbf.fit(X[0:40], data[0:40]).predict(-X[0:20])
pred = svr_rbf.fit(X[0:40], data[0:40]).predict(X[0:20])
pos_pred = svr_rbf.fit(X[0:40], data[0:40]).predict(X[20:40])
```

```python
plt.plot(-basis[0:20],neg_pred, 'b-_')
plt.plot(basis[0:20],pred, 'k')
plt.plot(basis[20:40],pos_pred, 'g-_')
plt.plot(basis,data, 'r-.')
plt.show()
```
